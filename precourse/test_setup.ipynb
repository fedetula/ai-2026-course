```python
import torch
print(f"PyTorch {torch.__version__} ... OK")
print(f"CUDA available: {torch.cuda.is_available()}")

# Quick autograd test
x = torch.tensor([2.0], requires_grad=True)
y = x ** 2
y.backward()
assert x.grad == 4.0, "Autograd failed"
print("Autograd ... OK")
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
plt.plot(x, np.sin(x))
plt.title("If you see this plot, you're all set!")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```
