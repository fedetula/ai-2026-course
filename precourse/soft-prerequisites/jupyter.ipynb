## 1. What Is Jupyter and Why ML People Use It

Jupyter Notebook (and its successor JupyterLab) is an interactive computing environment that lets you combine **executable code**, **rich text**, **plots**, and **equations** in a single document.

Why it matters for ML and data science:

- **Iterative exploration** -- you can run small pieces of code, inspect outputs, tweak, and re-run without restarting from scratch.
- **Inline visualization** -- plots appear right below the code that generated them.
- **Narrative + code** -- you can document your reasoning alongside the code, making analysis reproducible and shareable.
- **Industry standard** -- Kaggle, Google Colab, and most ML tutorials use notebooks.

The name "Jupyter" comes from the three core languages it originally supported: **Ju**lia, **Py**thon, **R**.

## 3. Running Cells

There are several ways to run a cell:

| Shortcut | Effect |
|----------|--------|
| **Shift + Enter** | Run the current cell and move to the next one |
| **Ctrl + Enter** | Run the current cell and stay on it |
| **Alt + Enter** | Run the current cell and insert a new cell below |

**Run order matters.** The number in brackets (e.g., `[3]`) to the left of a code cell shows the order in which it was executed. Python state depends on this order, not on the visual position of cells in the notebook.

````python
# Run this cell FIRST
x = 10
````

````python
# Then run this cell
print(f"x = {x}")
print(f"x * 2 = {x * 2}")
````

## 5. The Kernel

The **kernel** is the process that actually executes your code. For Python notebooks, it is a Python interpreter running in the background.

Key points:

- All code cells in a notebook **share the same kernel** (same memory, same variables).
- The kernel keeps running between cell executions -- variables persist.
- If something goes wrong (infinite loop, corrupted state), you can restart it.

### Kernel Operations (Kernel menu)

| Action | What it does |
|--------|-------------|
| **Interrupt** | Stops the currently running cell (like Ctrl+C) |
| **Restart** | Kills the kernel and starts fresh. All variables are lost. |
| **Restart & Run All** | Restarts and then runs every cell top to bottom. **Use this before sharing a notebook** to verify it runs cleanly. |
| **Restart & Clear Output** | Restarts and removes all cell outputs. |

**Best practice:** Before submitting or sharing any notebook, always do **Kernel > Restart & Run All**. If it does not run cleanly from top to bottom, something is broken.

````python
# Run this cell, then check the kernel indicator (top-right circle)
# While running, the circle is filled. When idle, it is empty.
import time
print("Starting...")
time.sleep(2)
print("Done! The kernel is idle again.")
````

## 6. Common Pitfalls

### Pitfall 1: Running cells out of order

The execution counter `[n]` shows when each cell was run. If you run cells out of order, you can get results that are impossible to reproduce.

````python
# Step 1: Run this cell
data = [1, 2, 3]
````

````python
# Step 2: Run this cell
data.append(4)
print(data)
````

If you run Step 2 **multiple times** without re-running Step 1, `data` keeps growing: `[1,2,3,4,4,4,...]`. This is a classic notebook bug.

### Pitfall 2: Hidden state

You might delete a cell that defined a variable, but the variable still exists in memory. The notebook looks correct, but after **Restart & Run All** it breaks because the definition is gone.

### Pitfall 3: Forgetting to re-run

You edit a function in cell 5 but forget to re-run it. Cells below still use the **old** version from memory. Always re-run a cell after editing it, and re-run any cells that depend on it.

## 7. Magic Commands

Jupyter provides special commands prefixed with `%` (line magic) or `%%` (cell magic). These are not Python -- they are Jupyter-specific.

| Magic | Scope | What it does |
|-------|-------|--------------|
| `%timeit` | Line | Times a single line (runs it many times for accuracy) |
| `%%time` | Cell | Times the entire cell (runs it once) |
| `%matplotlib inline` | Line | Makes matplotlib plots appear inline (default in modern Jupyter) |
| `%who` | Line | Lists all defined variables |
| `%whos` | Line | Lists variables with type and value info |
| `%reset` | Line | Clears all variables (like a soft kernel restart) |

````python
# %timeit -- benchmarks a single line
%timeit sum(range(1000))
````

````python
%%time
# %%time -- times the whole cell (single run)
total = 0
for i in range(1_000_000):
    total += i
print(f"Total: {total}")
````

````python
# %who -- list all variables currently in memory
a = 1
b = "hello"
c = [1, 2, 3]
%who
````

## 8. Keyboard Shortcuts Cheat Sheet

Jupyter has two modes:

- **Command mode** (blue border) -- press `Esc` to enter. Operates on cells.
- **Edit mode** (green border) -- press `Enter` to enter. Edits cell content.

### Command Mode (press Esc first)

| Shortcut | Action |
|----------|--------|
| `A` | Insert cell above |
| `B` | Insert cell below |
| `D, D` | Delete selected cell |
| `Z` | Undo cell deletion |
| `M` | Change cell to Markdown |
| `Y` | Change cell to Code |
| `Shift + Up/Down` | Select multiple cells |
| `Shift + M` | Merge selected cells |
| `L` | Toggle line numbers |
| `0, 0` | Restart kernel |
| `I, I` | Interrupt kernel |
| `H` | Show all shortcuts |

### Edit Mode (press Enter first)

| Shortcut | Action |
|----------|--------|
| `Ctrl + Shift + -` | Split cell at cursor |
| `Ctrl + /` | Toggle comment |
| `Tab` | Code completion / indent |
| `Shift + Tab` | Show docstring tooltip |

### Both Modes

| Shortcut | Action |
|----------|--------|
| `Shift + Enter` | Run cell, move to next |
| `Ctrl + Enter` | Run cell, stay |
| `Alt + Enter` | Run cell, insert below |
| `Ctrl + S` | Save notebook |
